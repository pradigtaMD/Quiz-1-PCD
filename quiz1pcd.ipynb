```python
from google.colab import files
import cv2
import numpy as np
from PIL import Image
import io
import matplotlib.pyplot as plt
```

```python
image_path = '/content/avatar.21.08.jpeg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
gaussian_noise = np.zeros(image_rgb.shape, dtype=np.uint8)
cv2.randn(gaussian_noise, 0, 25)  # Standar deviasi noise
noisy_image_gaussian = cv2.add(image_rgb, gaussian_noise)
```

```python
salt_pepper_noise = np.zeros(image_rgb.shape, dtype=np.uint8)
cv2.randu(salt_pepper_noise, 0, 255)  # Random noise
salt_pepper_noise[salt_pepper_noise < 30] = 0  # Salt
salt_pepper_noise[salt_pepper_noise > 225] = 255  # Pepper
noisy_image_salt_pepper = cv2.add(image_rgb, salt_pepper_noise)
```

```python
x = np.random.randint(0, width - 250)
y = np.random.randint(0, height - 250)
random_crop = gray_image[y:y + 250, x:x + 250]
```

```python
center_crop = gray_image[(height - 250) // 2:(height + 250) // 2, (width - 250) // 2:(width + 250) // 2]
```

```python
plt.subplot(3, 4, 1), plt.imshow(image_rgb), plt.title('Original')
plt.subplot(3, 4, 2), plt.imshow(gray_image, cmap='gray'), plt.title('Grayscale')
plt.subplot(3, 4, 3), plt.text(0.5, 0.5, f"Width: {width}, Height: {height}", fontsize=12, ha='center'), plt.axis('off')
plt.subplot(3, 4, 4), plt.text(0.5, 0.5, f"Max Pixel: {max_pixel}, Min Pixel: {min_pixel}", fontsize=12, ha='center'), plt.axis('off')
plt.subplot(3, 4, 5), plt.imshow(normalized_image, cmap='gray'), plt.title('Normalized')
plt.subplot(3, 4, 6), plt.imshow(flip_horizontal), plt.title('Flip Horizontal')
plt.subplot(3, 4, 7), plt.imshow(flip_vertical), plt.title('Flip Vertical')
plt.subplot(3, 4, 8), plt.imshow(noisy_image_gaussian), plt.title('Gaussian Noise')
plt.subplot(3, 4, 9), plt.imshow(noisy_image_salt_pepper), plt.title('Salt Pepper Noise')
plt.subplot(3, 4, 10), plt.imshow(random_crop, cmap='gray'), plt.title('Random Crop')
plt.subplot(3, 4, 11), plt.imshow(center_crop, cmap='gray'), plt.title('Center Crop')
```

```python
plt.tight_layout()
plt.show()
```

```python
print(f"Lebar gambar: {width} piksel")
print(f"Tinggi gambar: {height} piksel")
print(f"Nilai maksimum pixel: {max_pixel}")
print(f"Nilai minimum pixel: {min_pixel}")
```
